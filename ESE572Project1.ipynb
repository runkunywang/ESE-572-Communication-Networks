```
# Trie insert Code

class Node:
  def __init__(self, val):
    self.left = None
    self.right = None
    self.val = val

class Trie:
  def __init__(self):
    self.root = None

  def add(self, address, hopVal, length):
    # if trie root is empty then start it with an empty node
    if self.root == None:
      self.root = Node(None) 

    # creating counter for length
    depth = 0
    # starting from root of trie
    currentNode = self.root
    # for each number in the address
    for number in address:
      # if number is 0 in address go left
      # print(depth, number)
      if number == "0":
        # if the next left node doesnt exist
        if currentNode.left is None:
          # create a new node
          newNode = Node(None)
          # connecting it to current node
          currentNode.left = newNode
        # traversing down one node
        currentNode = currentNode.left
        # print("going left")
        depth += 1
        if depth == length:
          currentNode.val = hopVal
          return currentNode.val
      # if number is 1 in address go right
      if number == "1":
        # if the next right node doesnt exist
        if currentNode.right is None:
          # create a new node
          newNode = Node(None)
          currentNode.right = newNode
        currentNode = currentNode.right
        # print("going right")
        depth += 1
        if depth == length:
          currentNode.val = hopVal
          return currentNode.val
    return

  def lookup(self, address):
    # getting starting time
    startTime = time.time()
    # Wenxuan you can write your code here for searching address in the tree
    
    # get the length of adress
    num = len(address) 

    # store the closest hop
    hop = []
    
    # store largest matching prefix that has a hop
    pre = ''

    # store the remaining str after the pre
    rem = ''

    # store the tree depth
    tDep = 0;

    # initialize the tris
    T = self.root
    j = 0;

    # start from the first str of adress to the last one
    for i in range(num):

      # add the str to rem
      rem += address[i]

      # if 0 go the left, if 1 go the right, if has a non-none value store pre and hop
      if address[i] == '0':
        j += 1
        if T.left is None:
          break
        T = T.left
      elif address[i] == '1':
        j += 1
        if T.right is None:
          break
        T = T.right
      
      if T.val != None:
        hop = T.val
        pre += rem
        rem = ''
        tDep += j
        j = 0
    
    
    endTime = time.time()
    totalTime = endTime - startTime
    # print(pre, hop, totalTime, tree depth)
    return [address, pre, hop, tDep, round(totalTime,6)]
            
        
    

```

```
# idea for second data structure: using hash table. Create hash table with prefixs. 
# Searching: take the address and see if key yields value in hash table. If not then take last digit off and try again
# repeat until finds a key that works or returns that there is no key in the table for the address

class addressHash:
  def __init__(self):
    self.table = {}
  
  def addToTable(self, address, hopVal):
    self.table[address] = hopVal
    return
  
  def searchTable(self, address):
    iters = 0
    startTime = time.time()
    addArr = list(address)
    sep = ''
    while len(addArr) > 0:
      joinedAddress = sep.join(addArr)
      # print(joinedAddress)
      if joinedAddress in self.table:
        finishTime = time.time()
        totTime = finishTime - startTime
        return [address, joinedAddress, self.table[joinedAddress], iters, round(totTime,6)] 
      else:
        iters += 1
        addArr.pop()
    return 'not in table'
```
